# CAPTONE PROJECT - Taiwind Traders Report

### 📖 Project Overview

This project was completed as a part of the Microsoft Power BI Professional Certificate Capstone Project.
The goal was to design and deliver a comprehensive business analysis report for TAILWIND TRADERS, a fictional retail company, demonstrating the ability to transform raw data into actionable insights using Power BI.


### 🎯 Objective

- Import data from Excel and Python Script data sources into Power BI
- Clean and rectify any discrepancies or errors in Power Query
- Construct a snowflake schema tailored for the data model
- Build interactive dashboards with clear KPIs.
- Apply DAX measures and calculations for advanced analytics.
- Deliver insights into sales, product trends, and profit overview


### 🛠️ Tools & Technologies
- Excel/CSV datasets – Source data provided for Tailwind Traders.
- Power Query – Data cleaning and transformation.
- DAX (Data Analysis Expressions) – Custom measures and calculations.
- Power BI Desktop – Data modeling, visualization, and report building.


### 📊 Key Features of the Report
Tailwind Traders Global Sales and Profits

- Sales Overiew - Sales Trends, top-performing products, regional sales breakdown
- Profit Overview - Proft Margin and revenue analysis 
- Interactive Filters: Drill-down capabilities by  geography and date


### 💡 Business Insights
The analysis of Tailwind Traders’ data revealed several actionable insights:
• 	## 💡 Business Insights
- **Customer Loyalty**
  - UK leads with **315 loyalty points**, followed by USA with **305**.
  - Indicates strong engagement in these two markets compared to others.

- **Sales & Inventory**
  - Overall **median sales ~ $222**.
  - **152 total units sold**, while inventory remains high at **~14,000 units**.
  - No further stocking is needed; future stocking should be **demand-driven**.

- **Top Products**
  - **Floral Wall, Porcelain Dinner Set, and ProCarpenter Toolkit** are the **top 3 selling products**, each with **6 units sold**.
  - **Modular Sofa Set** generated the **highest revenue (~$928)** among all products.

- **Regional Performance**
  - **UAE has the highest median sales (~$680)**, showing stronger purchasing power compared to other regions.

- **Seasonal Trends**
  - **September 5, 2023 recorded the highest median sales ($998.24)** across the year.
  - Indicates potential seasonal demand spikes worth leveraging.

- **Revenue & Profitability**
  - **Total net revenue ~ $14,000**.
  - **Year-to-date profit margin (~62%)**, stable across the year except for a **sharp decline on October 1, 2023**, which quickly recovered.
  - **Country-level profit margins remain stable (~20%)**, showing consistent performance across regions.

## 📌 Recommendations
1. **Inventory Optimization**
   - Avoid overstocking; align future stocking with **demand forecasts**.
   - Implement **just-in-time inventory practices** to reduce holding costs.

2. **Market Strategy**
   - Focus marketing and loyalty programs on **UK and USA**, where customer engagement is already strong.
   - Expand premium product offerings in **UAE**, leveraging its higher median sales.

3. **Product Portfolio**
   - Prioritize promotion of **top-selling products** (Floral Wall, Porcelain Dinner Set, ProCarpenter Toolkit).
   - Explore bundling strategies or upselling with **Modular Sofa Set**, the highest revenue generator.

4. **Seasonal Campaigns**
   - Capitalize on **early September demand spikes** with targeted promotions and campaigns.
   - Use historical sales data to predict and prepare for similar seasonal surges.

5. **Profit Margin Monitoring**
   - Investigate the **October 1, 2023 margin dip** to identify root causes (e.g., discounts, supply chain issues).
   - Maintain continuous monitoring to ensure margins remain stable and sustainable.

6. **Customer Engagement**
   - Enhance loyalty programs in **UK and USA** to strengthen retention.
   - Introduce **personalized offers** for repeat customers, who already contribute significantly to revenue.
